```python

sn_total= purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
sn_avg= purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
sn_cuenta= purchase_data.groupby(["SN"]).count()["Price"].rename("Total Purchase Value")


concentrado = pd.DataFrame({"Purchase Count": sn_cuenta,
                            "Average Purchase Price":sn_avg,
                            "Total Purchase Value":sn_total
                                    })

concentrado1 = concentrado.sort_values("Purchase Count", ascending=False)

concentrado1["Average Purchase Price"] = concentrado1["Average Purchase Price"].map("${:.2f}".format)
concentrado1["Total Purchase Value"]   = concentrado1["Total Purchase Value"].map("${:.2f}".format)


concentrado1
```

```python
# Hacemos los data frames de cada requerimiento
Item_total= purchase_data.groupby(["Item ID"]).sum()["Price"].rename("Total Purchase Value")
Item_Price= purchase_data.groupby(["Item ID"]).mean()["Price"].rename("Item Price")
Item_cuenta= purchase_data.groupby(["Item ID"]).count()["Item Name"].rename("Purchase Count")
# los juntamos en un sólo data frame
concentrado2 = pd.DataFrame({"Purchase Count": Item_cuenta,
                            "Item Price":Item_Price,
                            "Total Purchase Value":Item_total
                                    })
# lo ordenamos
concentrado3 = concentrado2.sort_values("Purchase Count", ascending=False)
# lo formateamos
concentrado3["Item Price"] = concentrado3["Item Price"].map("${:.2f}".format)
concentrado3["Total Purchase Value"]   = concentrado3["Total Purchase Value"].map("${:.2f}".format)
# hacemos el data frame de nombres
items=purchase_data[ ["Item ID","Item Name"]]
# quitamos los dupliocaod
items = items.drop_duplicates()
#junatmos las tablas
merge_table = pd.merge(concentrado3, items, on="Item ID")
merge_table1=merge_table[["Item ID","Item Name","Purchase Count","Item Price","Total Purchase Value"]]
merge_table1
```
